```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Create a mapping
day_files = {
    'Monday':    'TRAINING_DATA_MON.csv',
    'Tuesday':   'TRAINING_DATA_TUE.csv',
    'Wednesday': 'TRAINING_DATA_WED.csv',
    'Thursday':  'TRAINING_DATA_THU.csv',
    'Friday':    'TRAINING_DATA_FRI.csv',
    'Saturday':  'TRAINING_DATA_SAT.csv',
    'Sunday':    'TRAINING_DATA_SUN.csv'
}

dependent_var = 'raw_MM_Wh'
independent_vars = ['D-3', 'D-2', 'D-1', 'D+1', 'D+2', 'D+3']

# Regression analysis and plotting of data for each day
for day, file_path in day_files.items():
    df = pd.read_csv(file_path)

# Construct a regression model
    Y = df[dependent_var]
    X = df[independent_vars]
    X = sm.add_constant(X)
    model = sm.OLS(Y, X).fit()
    y_pred = model.predict(X)
    
# Print summary
    print(f"\n{'='*40}\nregression resultsï¼š{day}\n{'='*40}")
    print(model.summary())
    
# Plot the actual values against the forecast
    plt.figure(figsize=(12, 6))
    if 'dateTimeStmp' in df.columns:
        plt.plot(df['dateTimeStmp'], Y, label='Actual (raw_MM_Wh)', color='blue')
        plt.plot(df['dateTimeStmp'], y_pred, label='Predicted', color='red')
        plt.xticks(rotation=45) 

        # Set the x-axis scale manually: Only the years 2023 and 2024 are displayed
        tick_positions = []
        tick_labels = []
        for year in ['2023', '2024']:
            matching_indices = df.index[df['dateTimeStmp'].str.startswith(year)]
            if not matching_indices.empty:
                tick_positions.append(matching_indices[0])
                tick_labels.append(year)
        plt.xticks(tick_positions, tick_labels, rotation=45)
        
    else:
        plt.plot(Y.index, Y, label='Actual (raw_MM_Wh)', color='blue')
        plt.plot(Y.index, y_pred, label='Predicted', color='red')
    
    plt.xlabel('Time')
    plt.ylabel('raw_MM_Wh')
    plt.title(f'Actual vs. Predicted raw_MM_Wh ({day})')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
